```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load and explore data
# Load dataset
df = pd.read_csv('asian_games_medals.csv')

# Preview the data
df.head()

# Filter Malaysia Data
# Look at unique countries to find Malaysia's exact name
print(df['NOC'].unique())

# Filter for Malaysia
df['NOC'] = df['NOC'].apply(
    lambda x: 'Malaysia' if 'Malaysia' in x or 'Malaya' in x else x
)
df_malaysia = df[df['NOC'] == 'Malaysia'].copy()

# Reset index
df_malaysia = df_malaysia.reset_index(drop=True)
df_malaysia.head()

# Clean and prepare
# Check for missing values
df_malaysia.isnull().sum()

# Make sure Year is sorted
df_malaysia = df_malaysia.sort_values('Year')

# Add total medal column if not present
df_malaysia['Total'] = df_malaysia['Gold'] + df_malaysia['Silver'] + df_malaysia['Bronze']

# Thinking Direction / Question to Answer
# How many medals has Malaysia won in each Asian Games edition?

# What is the trend of total medals over time?
# Total Medal Trend Over Time
plt.figure(figsize=(10,6))
sns.lineplot(data=df_malaysia, x='Year', y='Total', marker='o')
plt.title('Malaysia Total Medals Over Asian Games Editions')
plt.ylabel('Total Medals')
plt.xlabel('Year')
plt.grid(True)
plt.show() 

# Which year was Malaysias best performance?

# How does gold medal count compare to silver and bronze over time?
# Gold vs Silver vs Bronze Comparison
plt.figure(figsize=(10,6))
sns.lineplot(data=df_malaysia, x='Year', y='Gold', label='Gold', marker='o')
sns.lineplot(data=df_malaysia, x='Year', y='Silver', label='Silver', marker='o')
sns.lineplot(data=df_malaysia, x='Year', y='Bronze', label='Bronze', marker='o')
plt.title('Malaysia Medal Breakdown (Gold, Silver, Bronze)')
plt.ylabel('Medals')
plt.xlabel('Year')
plt.legend()
plt.grid(True)
plt.show()

print(df_malaysia.shape)

# Show all rows for Malaysia
print(df_malaysia)



```

#  Malaysia at the Asian Games: Insights

- The country's **peak performance** was in **2006** with **42**.
- Over time, Malaysias **total medal count shows steady growth**, especially from the 1990s onwards.
- Malaysia tends to win **more bronze and silver medals** than gold, indicating strong regional competitiveness.
